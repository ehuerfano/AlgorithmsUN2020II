```
from itertools import permutations  
import matplotlib.pyplot as plt
import math 
from collections import Counter
```

```
def swap (v, i, j):
	t = v[i]
	v[i] = v[j]
	v[j] = t

def bubbleSort(v):
      pasos = 0
      cadena = [v[i] for i in range(len(v))]
  
      for i in range (len(v)):
          swaps = 0
          for j in range (len(v)-1,i,-1):
              pasos+=1
              if (cadena[j] < cadena[j-1]):
                  swap(cadena,j,j-1)
                  pasos+=1
                  swaps+=1
          if (swaps==0):
            break;
      return pasos
```

```
def paint(n):
		
    v = [i+1 for i in range(n)] 
    stps = []
    
    frec = [0 for k in range(400000)] 

    permutaciones = permutations(v)

    for i in permutaciones: 
        k = bubbleSort(list(i))
        stps.append(k)

        frec[k] += 1

    permutations_count = math.factorial(n)
    print("\nn = ",n)
    print("Number of permutations = ",permutations_count)
    print("Average number of steps = ",round(sum(stps)/len(stps),1))
    occurrences = Counter(stps)
    for key in occurrences:
        print("P("+str(key)+") = "+str(occurrences[key]/permutations_count))
    plt.hist(stps,  
          bins = [i for i in range(int(min(stps)),int(max(stps))+2,1)],
          # alpha=1,
          density=True,
          histtype='bar',
          edgecolor='k')
    plt.show()
```

```
for n in range (3,10):
  paint(n)
```
