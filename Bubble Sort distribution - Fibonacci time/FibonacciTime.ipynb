```
import matplotlib.pyplot as plt
```

```
def fibR(n):
    if n==0 :
        return 0
    if n==1 :
        return 1
    return fibR(n-1)+fibR(n-2)
```

```
import time
fibR_time = []
fibR_comp = []
for n in range(1,45):
    t0 = time.clock()
    k = fibR(n)
    tf = time.clock()
    fibR_time.append(tf - t0)
    fibR_comp.append(n)
plt.plot(fibR_time)
plt.plot(fibR_comp)
plt.show()
```

```
def fibI(n):
    if n==0 :
        return 0
    a, b = 0, 1
    for i in range(1, n):
        a, b = b, a+b
    return b
```

```
import time
fibI_time = []
fibI_comp = []
for n in range(1,501):
    t0 = time.clock()
    k = fibI(n)
    tf =time.clock()
    fibI_time.append(tf - t0)
    fibI_comp.append(n/10000000)
plt.plot(fibI_time)
plt.plot(fibI_comp)
plt.show()
```
